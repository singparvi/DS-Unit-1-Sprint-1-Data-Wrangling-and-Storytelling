###2) Create the visualizations you selected in 2).  

```python
### YOUR CODE HERE 
import matplotlib.pyplot as plt
import pandas as pd
# using the pyplot API approach 

# fig, ax = plt.subplots(figsize = (8,8))

plt.hist(Titanic['Fare'], label = 'Titanic Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Observations')
plt.title('Titanic Fare Histogram')
plt.show()
```

###3) Describe your results in a sentence or two.

Answer: The histogram is skewed to the right and is unimodal and asymmetrical. Gererally speaking it can be inferred that the majority of people paid between 0-50 for the Titanic Fare.

###4) Plot a histogram of fare paid by Titanic passengers by sex to see if there was a difference in fare paid between male and female passengers.  

```python
### YOUR CODE HERE 
condition_male = Titanic['Sex'] == 'male'
condition_female = Titanic['Sex'] == 'female'

# assigning dataframe by condition
Titanic_Male = Titanic[condition_male]
Titanic_Female = Titanic[condition_female]
# plotting Male
plt.hist(Titanic_Male['Fare'])
plt.xlabel('Fare')
plt.ylabel('Number of Observations')
plt.title('Titanic Fare distribution for Male')
plt.show()
# plotting Female
plt.hist(Titanic_Female['Fare'])
plt.xlabel('Fare')
plt.ylabel('Number of Observations')
plt.title('Titanic Fare distribution for Female')
plt.show()
```

###5) Describe the difference in distribution of fare paid between male and female passengers.

Answer:- The difference in the distribution of fare paid between male and female passengers is very minimilistic. From this data it can be inferred that the population of male passengers were more than female passengers and both gender group generally paid the same range of fare with both data unimodal and skewed to the right.

###In 10) - 14)  We will use figure, axes syntax to plot passenger survival by whether the passenger was a child or an adult. 

###10) Create a variable to categorize the passenger as a child (under age 18) or adult (18 years or older).

```python
### YOUR CODE HERE 

df = Titanic.copy()
df.loc[df['Age'] >= 18, 'Adult/Child'] = 'Adult'
df.loc[df['Age'] < 18, 'Adult/Child'] = 'Child'
df
```

###11) Make a cross-tab of survival by age (child vs adult).

```python
### YOUR CODE HERE 
pd.crosstab(index = df['Survived'], columns = df['Adult/Child'])
```

###12) Create a new data frame to hold the data for the figure.

```python
### YOUR CODE HERE 

df1 = pd.DataFrame({'Adult': [480, 277], 'Child': [65, 65]})
df1
```

###13) Adapt the plot from the Guided Project to show the relationship between passenger age (child vs adult) and survival.

```python
### YOUR CODE HERE 
fig, ax = plt.subplots(figsize = (8, 8))

my_ticks = ['Did not survide', 'Survived']
ax.plot(my_ticks, df1['Adult'], label = 'Adult')
ax.plot(my_ticks, df1['Child'], label = 'Child')
ax.set_title('Survival rate of Adult vs Child on Titanic')
ax.set_xlabel('Passenger Survival')
ax.set_ylabel('Number of Passengers')
ax.legend()
ax.plot()
```
