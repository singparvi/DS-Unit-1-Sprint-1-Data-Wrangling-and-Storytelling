Objective 02 - use appropriate terminology when referring to parts of a Matplotlib graph

https://matplotlib.org/3.2.1/gallery/showcase/anatomy.html

```
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```
# Generate the X and multiple Y values
X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len (X))
Y3
# create a figure object with a size of 8X 8 inches
fig = plt.figure(figsize= (8,8))
# creat the figures axes with a single plot
ax = fig.add_subplot(1,1,1)
# Finally plot
# we are choosing different colors here

# plot a solid magenta line, width = 2
ax.plot(X, Y1, c = 'm', lw = 2, label = 'magenta signal')

# plot an orange dash-dot line, with = 2
ax.plot(X, Y2, c = "orange", lw = 3, linestyle = '--', label = 'orange signal')

# plot a yellow circle marker with a green edge, line width = 0 (not visible)
ax.plot(X, Y3, linewidth = 0, marker ='o', markerfacecolor = 'y', markeredgecolor = 'g')

# set the tile and x-y labels
ax.set_title("Anatomy of a figure (modified)", fontsize = 18)
ax.set_xlabel("X Axis Label")
ax.set_ylabel('Y Axis Label')
ax.legend()

plt.show()
```

Objective 03 - differentiate between Matplotlib syntaxes - pyplot and object-oriented


```
# Create the x-axis data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# create a plot figure
fig = plt.figure(figsize= (8, 8))

# create a subplot which will on the top with 2 rows, 1 column and 1 position
ax = fig.add_subplot(2,1,1)
ax.plot(x, y1, color = 'b')

ax = fig.add_subplot(2,1,2)
ax.plot(x, y2, color = 'orange')

# adding a title to the plot
plt.suptitle('Pyplot Interface Example')
plt.plot()
```

```
# use the same data as above 

# create a grid of plots (notice the 's' in subplots)
# fig, ax - two axes objects
fig, ax = plt.subplots(2, figsize = (8,8))
# Look at the size of ax
print ('Number of ax created:', ax.size)

# call the plot() method on each axes object
# object 1
ax[0].plot(x, y1, 'b')

# object 2
ax[1].plot(x, y2, color = 'orange')

# addd a title to the top axes object
ax[0].set_title('Object-oriented Interface Example');

# fig.clf() # clear the figure
```

Objective 04 - use Matplotlib and stylesheets to control basic visual aspects of a plot


```
import matplotlib.pyplot as plt

print (plt.style.available)
```

```
# import the libraries

# set the style-sheet

plt.style.use('seaborn-bright')

# creating x variable
x = np.linspace(0,10)

# create figure and axes object
fig, ax = plt.subplots(figsize = (8,8))

# plot the random data
ax.plot(x, np.sin(x) + x + np.random.randn(50))
ax.plot(x, np.sin(x) + 0.5 * x + np.random.randn(50))
ax.plot(x, np.sin(x) + 2 * x + np.random.randn(50))
ax.plot(x, np.sin(x) - 0.5 * x + np.random.randn(50))
ax.plot(x, np.sin(x) + np.random.randn(50))

ax.set_title("'seaborn-bright' style sheet")
# plt.clf()
```
